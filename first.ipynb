```python
def get_HCF(a,b):
    '''
    This is to print HCF of 2 nos
    '''
    hcf=1
    s=a
    if s>b:
        s=b
    
    for i in range (1,s+1):
        if (a%i==0) and (b%i==0):
            hcf =i
    return hcf
    
n1=int(input("enter a number "))
n2=int(input("enter a number "))


print("HCF of {0} and {1} is {2}".format(n1,n2,get_HCF(n1,n2)))
        
```

```python
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
my_url="https://www.imdb.com/search/title/?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012"
uclient=ureq(my_url)
page_html=uclient.read()
uclient.close()  #The close() method closes an open file. You should always close your files, in some cases, due to buffering, changes made to a file may not show until you close the file.
```

```python
page_soup=soup(page_html) #html parser
```

```python
filename=("Movies_IMDB.csv")
f=open(filename,"w")

header="Name, Year, Runtime \n"
f.write(header)
```

```python
containers=page_soup.findAll("div", {"class": "lister-item mode-advanced"})
containers
```

```python
filename=("Movies_IMDB.csv")
f=open(filename,"w")

header="Name, Year, Runtime \n"
f.write(header)

for i in containers:
    name=i.img["alt"]
    year_movie=i.findAll("span", {"class": "lister-item-year"})
    year=year_movie[0].text
    runtime=i.findAll("span", {"class": "runtime"})
    runtime=runtime[0].text
    print(name+ "," + year +"," + runtime +"\n")
    f.write(name+ "," + year +"," + runtime + "\n")

f.close()
```
