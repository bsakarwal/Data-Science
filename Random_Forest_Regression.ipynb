```python
y_train=df_train['price']
x_train=df_train.drop('price',axis=1)
```

```python
x_train.head()
```

### Import Sklearn Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(max_depth=5,n_jobs=-1,min_samples_leaf=10,random_state=42)
```

```python
rf.fit(x_train,y_train)
```

```python
sample_tree=rf.estimators_[11]
```

```python
from IPython.display import Image  
from six import StringIO  
from sklearn.tree import export_graphviz
import pydotplus, graphviz
```

```python
dot_data = StringIO() 

export_graphviz(sample_tree, out_file=dot_data,
                feature_names=x_train.columns,filled=True,rounded=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
y_pred_train=rf.predict(x_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y_pred_train,y_train)
```

```python
mean_squared_error(y_pred_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={'max_depth':range(6,14,2),'min_samples_leaf':range(5,15,5),'n_estimators':range(10,100,10)}
```

```python
grid_search=GridSearchCV(estimator=rf,param_grid=params,n_jobs=-1,verbose=1,cv=4,scoring="accuracy")
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_est=grid_search.best_estimator_
```

```python
best_est
```

```python
best_est.fit(x_train,y_train)
```

```python
y_pred_train=best_est.predict(x_train)  # for train dataset
```

```python
print(r2_score(y_pred_train,y_train))
```

```python
best_est.feature_importances_
```
