```python
from sklearn.model_selection import train_test_split
```

```python
rf=RandomForestClassifier(max_depth=3,n_estimators=10,random_state=0)
```

```python
rf.fit(x_train,y_train)
```

```python
rf.get_params()
```

```python
l=[rf.get_params()]
#df=pd.DataFrame(rf.get_params())
l=pd.DataFrame(l)
l
```

```python
decision_tree=rf.estimators_
```

```python
lst=[]
row_label=[]
for idx,tree in enumerate (decision_tree):
    print(f"Decision tree {idx+1} parameters:")
    #df=pd.DataFrame(tree.get_params())
    row_label.append(f"Decision tree {idx+1} parameters:")
    lst.append(tree.get_params())
    #print(tree.get_params())
```

```python
row_label
```

#### Hyper Parameter Tunning

```python
classifer_rf=RandomForestClassifier(random_state=42,n_jobs=-1)

```

```python
classifer_rf.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={'max_depth':[3,4,5,10,20],'min_samples_leaf':[5,10,20,50,100],'max_features':[2,3,4],'n_estimators':[10,30,50,100,200]}
```

```python
grid_search=GridSearchCV(estimator=classifer_rf,param_grid=params,n_jobs=-1,verbose=1,cv=4,scoring="accuracy")
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.cv_results_
```

```python
grid_search.best_score_
```

```python
from sklearn.metrics import roc_curve
```

```python
roc_curve(y_pred_train,y_train)
plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(y_train, y_pred_train)
plt.plot(fpr, tpr)
plt.show()
```

```python
best_est=grid_search.best_estimator_
```

```python
best_est
```

```python
best_est.fit(x_train,y_train)
```

```python
y_pred_train=best_est.predict(x_train)  # for train dataset
```

```python
print(classification_report(y_pred_train,y_train))
```

```python
y_pred_test=best_est.predict(x_test)
```

```python
print(classification_report(y_pred_test,y_test))
```

```python
def evaluate_model(classifier):
    print("Train Accuracy :", accuracy_score(y_train, classifier.predict(x_train)))
    print("Train Confusion Matrix:")
    print(confusion_matrix(y_train, classifier.predict(x_train)))
    print("-"*50)
    print("Test Accuracy :", accuracy_score(y_test, classifier.predict(x_test)))
    print("Test Confusion Matrix:")
    print(confusion_matrix(y_test, classifier.predict(x_test)))
```

```python
evaluate_model(best_est)
```
