```python
supplier=pd.read_csv('supplier_data.csv')
```

```python
supplier   # 0 rating means as per the feature and 0.5 is not in favour of feature eg if product is available then its 0 if less available/risky then its 0.5
```

```python
supplier.columns
```

```python
supplier['risk index']=supplier['availability']+supplier['no_suppliers']+supplier['standard']+supplier['price_fluctuation']
```

```python
supplier['Value']=supplier['Quantity']*supplier['price']
```

```python
supplier.head()
```

```python
supplier.describe()
```

```python
3.011452e+06
```

```python
def supplier_matrix(x,y):
    if ((x>=4000000) & (y>=1)):
        return 'Strategic'
    elif((x>=4000000) & (y<1)):
        return 'Leverage'
    elif((x<4000000) & (y<1)):
        return 'Non-Critical'
    elif((x<4000000) & (y>=1)):
        return 'Bottel Neck'
```

```python
supplier_matrix(240000,2)
```

```python
#supplier['Category']=supplier[['Value','risk index']].apply(supplier_matrix)
supplier['Category']=supplier.apply(lambda x: supplier_matrix(x['Value'], x['risk index']),axis=1)
```

```python
supplier.head()
```

```python
supplier['Category'].value_counts()
```

```python
supplier[supplier['Category']=='Leverage']
```

```python
sns.scatterplot(x='risk index',y='Value',data=supplier,hue='Category')
plt.axvline(x=1,linestyle='dotted', color='black')
plt.axhline(y=6000000, linestyle='dotted', color='black')
plt.plot()
```
